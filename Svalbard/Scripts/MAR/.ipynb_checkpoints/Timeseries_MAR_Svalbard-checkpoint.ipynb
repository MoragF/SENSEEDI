```python
import xarray as xr
import numpy as np
import scipy.io as io
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from affine import Affine
tif = xr.open_rasterio('/home/s1423313/Documents/Svalbard/Regions/Polys/07_rgi60_Svalbard_corr_land_R1.tif')

ds= xr.open_dataset('MARv3.11.2-6km-daily-ERA5-2000.nc')
#Fixing coordinates to lat and lon in prep for attempting to use mask
ds2=ds.set_coords(['LAT','LON'])
ds2.coords['y'] = tif.y.values
ds2.coords['x'] = tif.x.values
ds2.coords['band'] = tif.band.values
ds2.coords['mask']= (('band','y','x'),tif)


LRx1=[]
LRy1=[]
lon=[]
lat=[]
g=min(ds2.y)
h=max(ds2.y)
k=min(ds2.x)
l=max(ds2.x)
for i in range(len(ds2.Y20_218.values)):
    for j in range(len(ds2.X20_340.values)):
        a=ds2.LAT[i][j].values
        b=ds2.LON[i][j].values
        if g<=a<=h and k<=b<=l:
            e=min(ds2.y.values, key=lambda n:abs(n-a))
            f=min(ds2.x.values, key=lambda n:abs(n-b))
            lat.append(a)
            lon.append(b)
            c=list(ds2.y.values).index(e)
            d=list(ds2.x.values).index(f)
            if ds2.mask[0][c][d]==1:
                LRy1.append(ds2.Y20_218[i].values)
                LRx1.append(ds2.X20_340[j].values)
                    
np.save('LRx1',LRx1)
np.save('LRy1',LRy1)
```

```python
ds3= xr.open_mfdataset('/exports/csce/datastore/geos/groups/geos_EO/Databases/MAR/Svalbard-RA/MARv3.11.*.nc')
```

```python
SMB_LR1 = ds3.SMB.sel(Y20_218=LRy1, X20_340=LRx1).groupby('TIME.day').mean(dim=('Y20_218','X20_340'))
SF_LR1 = ds3.SF.sel(Y20_218=LRy1, X20_340=LRx1).groupby('TIME.day').mean(dim=('Y20_218','X20_340'))
land_LR1=io.loadmat('/home/s1423313/Documents/Svalbard/ElevationData/tmsr_734338_737938_45_90_07_rgi60_Svalbard_corr_land_R1.mat')
land_elevation_LR1 = land_LR1['Tmsr'][0]
time = land_LR1['time'][0]
timepy= pd.to_datetime(time-719529, unit='D')
E_LR1 = xr.Dataset(data_vars = dict(elevation_land_LR1 = (["time"], land_elevation_LR1)),coords=dict(time=timepy))
```

```python

labels = ['Elevation Land', 'MAR: SMB','MAR: SF']
fig, (ax1, ax2, ax3) = plt.subplots(3,figsize=(18,10), sharex=True)
plt.suptitle('Land Terminating R1', fontsize=16)

color = 'tab:red'
ax1.set_ylabel('elevation[m]', color=color, fontsize=14)
ax1.plot(E_LR1.time, E_LR1.elevation_land_LR1, color=color)

color = 'tab:blue'
ax3.set_ylabel('mmWE/day', color=color, fontsize=14)  # we already handled the x-label with ax1
ax3.plot(SMB_LR1.TIME.sel(TIME=slice('2010-07-18','2019-12-31')), SMB_LR1.sel(TIME=slice('2010-07-18','2019-12-31')), color=color)

color = 'tab:green'
ax2.set_ylabel('mmWE/day', color=color, fontsize=14)  
ax2.plot(SF_LR1.TIME.sel(TIME=slice('2010-07-18','2019-12-31')), SF_LR1.sel(TIME=slice('2010-07-18','2019-12-31')), color=color)

fig.legend(labels)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])  # otherwise the right y-label is slightly clipped
plt.show()
```
